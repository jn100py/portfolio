### Deel II

- Een aantal gemeenten moeten qua naam gelijk getrokken worden

```python
for name, regiocode, new_name in [
                ['\'s-Gravenhage', 'G0518', 'Den Haag'],
                ['Bergen', 'G0373', 'Bergen (NH.)'],
                ['Bergen', 'G0893', 'Bergen (L.)'],
                      ]:

    filter1 = election_data['Gemeente'].isin([name])
    filter2 = election_data['RegioCode'].isin([regiocode])
    index = election_data[filter1][filter2].index
    election_data.loc[index, 'Gemeente'] = new_name
```

- De bijzondere gemeenten komen te vervallen in election_data

```python
election_data = election_data.drop(election_data[(election_data.Gemeente.isin(['Bonaire', 'Saba', 'Sint Eustatius']))].index)
```

- Nu het probleem oplossen met de samengevoegde gemeenten:<br>
In election_data dienen<br>
    * de nieuwe gemeentenamen toegevoegd te worden<br>
    * de uitgefaseerde gemeentenamen verwijderd te worden<br>
    * de aantallen van de uitgefaseerde gemeenten toegevoegd te worden aan de gemeente waarin ze werden opgenomen

```python
conversion_table = {
    'Dijk en Waard': ['Heerhugowaard', 'Langedijk'],
    'Land van Cuijk': ['Boxmeer', 'Cuijk', 'Grave', 'Mill en Sint Hubert', 'Sint Anthonis'],
    'Maashorst': ['Landerd', 'Uden'],
    'Purmerend': ['Purmerend', 'Beemster'],
    'Amsterdam': ['Amsterdam', 'Weesp'],
}
```

```python
def find_new_municipalities(conversion_table):
    return [municipality for municipality in conversion_table if municipality not in conversion_table[municipality]]


def find_updated_municipalities(conversion_table):
    return [municipality for municipality in conversion_table if municipality in conversion_table[municipality]]


def find_phased_out_municipalities(conversion_table, updated_municipalities):

    phased_out_municipalities = []

    for municipality_ in conversion_table:
        for municipality in conversion_table[municipality_]:
            if municipality not in updated_municipalities:
                phased_out_municipalities.append(municipality)

    return phased_out_municipalities


def add_new_municipalities(new_municipalities, conversion_table, election_data):
    
    for new_municipality in new_municipalities:
        phased_out_municipalities = conversion_table[new_municipality]
    
        municipality_to_add = election_data[(election_data.Gemeente.isin(phased_out_municipalities))].sum()
        municipality_to_add.Gemeente = new_municipality
        election_data = election_data.append(municipality_to_add, ignore_index=True)

        indices_rows2drop = election_data[(election_data.Gemeente.isin(phased_out_municipalities))].index
        election_data = election_data.drop(indices_rows2drop)

    return election_data


def update_municipalities(updated_municipalities, conversion_table, election_data):
    
    for updated_municipality in updated_municipalities:
        
        municipalities2merge = conversion_table[updated_municipality]

        municipality_to_add = election_data[(election_data.Gemeente.isin(municipalities2merge))].sum()
        election_data = election_data.drop(election_data[(election_data.Gemeente.isin(municipalities2merge))].index)
        municipality_to_add.Gemeente = updated_municipality
        election_data = election_data.append(municipality_to_add, ignore_index=True)
    
    return election_data


new_municipalities = find_new_municipalities(conversion_table)
updated_municipalities = find_updated_municipalities(conversion_table)
phased_out_municipalities = find_phased_out_municipalities(conversion_table, updated_municipalities)
    
election_data = add_new_municipalities(new_municipalities, conversion_table, election_data)
election_data = update_municipalities(updated_municipalities, conversion_table, election_data)
```
